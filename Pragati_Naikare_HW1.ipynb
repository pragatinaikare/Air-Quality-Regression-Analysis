## c) (20 points) Linear Regression Implementation

##### Implement a linear regression model from scratch to regress the target variable, Carbon monoxide (CO). (Remember: You can not use any libraries for the linear regression model.)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from math import sqrt

class LinearRegression:
    def __init__(self, alpha= 0.0005, iterations=25000):
        self.alpha = alpha
        self.iterations = iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        X = np.c_[np.ones(X.shape[0]), X]

        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for i in range(self.iterations):
            pred = self.predict(X)

            error = pred - y

            self.weights -= self.alpha * (1/X.shape[0]) * np.dot(X.T, error)
            self.bias -= self.alpha * (1/X.shape[0]) * np.sum(error)

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
    
```

### (d) (20 points) Logistic Regression Implementation

Using the column PT08.S1(CO), create a binary label for this dataset where the values more than 1000 correspond to label 1 
and the values less than or equal to 1000 correspond to label 0. Implement a logistic regression model from scratch to predict
this binary label. (Remember: You can not use any libraries for the logistic regression model.)

```python
class LogisticRegression:
    def __init__(self, alpha=0.0006, iterations=25000, grad_threshold=5.0):
        self.alpha = alpha
        self.iterations = iterations
        self.grad_threshold = grad_threshold
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        z = np.clip(z, -self.grad_threshold, self.grad_threshold)
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        total_samples, total_features = X.shape

        self.weights = np.zeros(total_features)
        self.bias = 0

        for epoch in range(self.iterations):
            lr_pred = self.sigmoid(np.dot(X, self.weights) + self.bias)

            w = np.dot(X.T, (lr_pred - y)) / total_samples
            b = np.sum(lr_pred - y) / total_samples

            dw = np.clip(w, -self.grad_threshold, self.grad_threshold)
            db = np.clip(b, -self.grad_threshold, self.grad_threshold)

            self.weights -= self.alpha * dw
            self.bias -= self.alpha * db

    def predict(self, X):
        return np.where(self.sigmoid(np.dot(X, self.weights) + self.bias) > 0.5, 1, 0)
```

### (e) (10 points) Result Analysis - Linear Regression 

##### Perform a 5-fold cross validation. Compute RMSE for each validation set across 5 folds. Report average and standard deviation of RMSE values. Do you see a big change across different folds? How can you use the coefficient of this model to find the most informative features?

```python

label_train = label_train.squeeze()
rmse_values = []
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_index, val_index) in enumerate(kf.split(normalized_data), 1):
    X_train, X_val = normalized_data.iloc[train_index], normalized_data.iloc[val_index]
    y_train, y_val = label_train.iloc[train_index], label_train.iloc[val_index]

    linear_model = LinearRegression()
    linear_model.fit(X_train, y_train)

    y_pred = linear_model.predict(np.c_[np.ones(X_val.shape[0]), X_val])

    rmse = sqrt(mean_squared_error(y_val, y_pred))
    rmse_values.append(rmse)

    print(f"RMSE for Fold {fold}: {rmse}")

avg_rmse = np.mean(rmse_values)
std_rmse = np.std(rmse_values)

print("\nAvg RMSE across 5 folds:", avg_rmse)
print("Standard Deviation of RMSE - 5 folds:", std_rmse)
```

Observing the RMSE values across different folds, we can see that there is some variation in the RMSE values, although it's not very large. The RMSE values range from approximately 69.93 to 74.26, with an average RMSE of around  71.95 and a standard deviation of approximately 1.80. This indicates that the model's performance is relatively consistent across different folds, with some variations due to differences in the data splits.



##### To answer the given question -  How can you use the coefficient of this model to find the most informative features? let's print the coeficient of each feature and analyze the output 

```python
linear_model = LinearRegression()
linear_model.fit(normalized_data, label_train)

print("Coefficients of the features:")
for i, coef in enumerate(linear_model.weights[1:], start=1):
    feature_name = normalized_data.columns[i - 1]  # Subtracting 1 because bias term is added
    print(f"{feature_name}: {coef}")

# Finding the most informative features
most_informative_features = np.argsort(np.abs(linear_model.weights[1:]))[::-1]

print("\nMost informative features (top 5):")
for i in range(5):
    feature_index = most_informative_features[i]
    feature_name = normalized_data.columns[feature_index]
    print(f"{feature_name}: {linear_model.weights[feature_index + 1]}")
```

Features with larger magnitude coefficients are considered more informative. In your output, features like "PT08.S5(O3)" and "PT08.S2(NMHC)" have the largest coefficients, indicating that they have a significant impact on the target variable.<br>
A positive coefficient suggests that as the feature value increases, the target variable also tends to increase. On the other hand, a negative coefficient suggests that as the feature value increases, the target variable tends to decrease. ### (f ) (10 points) Result Analysis - Logistic Regression Perform a 5-fold cross validation.
##### Compute accuracy, precision, recall, and F1 score for each validation set across 5 folds. Report the average and standard deviation of these metrics. Do you see a big change across different folds?

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


logistic_model = LogisticRegression()

data_combined = np.column_stack((normalized_data, label_train))

X_logistic = data_combined[:, :-1]
y_logistic = data_combined[:, -1]
y_logistic = (y_logistic > 1000).astype(int)

# Defining 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

fold = 1
for train_index, test_index in kf.split(X_logistic):
    X_train, X_val = X_logistic[train_index], X_logistic[test_index]
    y_train, y_val = y_logistic[train_index], y_logistic[test_index]

    logistic_model.fit(X_train, y_train)

    predictions = logistic_model.predict(X_val)

    accuracy = accuracy_score(y_val, predictions)
    precision = precision_score(y_val, predictions)
    recall = recall_score(y_val, predictions)
    f1 = f1_score(y_val, predictions)

    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

    print(f"Fold {fold} - Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")
    fold += 1
```

##### Do you see a big change across different folds?
Looking at the performance metrics across the different folds, there are some variations, but overall, the changes are not too drastic. 

```python
print("\nAverage Accuracy: ", np.mean(accuracy_scores))
print("Standard Deviation Accuracy: ", np.std(accuracy_scores))
print("Average Precision: ", np.mean(precision_scores))
print("Standard Deviation Precision: ", np.std(precision_scores))
print("Average Recall: ", np.mean(recall_scores))
print("Standard Deviation Recall: ", np.std(recall_scores))
print("Average F1 Score: ", np.mean(f1_scores))
print("Standard Deviation F1 Score: ", np.std(f1_scores))
```

### (g) (10 points) ROC Curve - Logistic Regression
##### Use the logistic regression model from sklearn and repeat 5-fold cross validation.  Then using roc curve package from sklearn.metrics, plot the ROC curve for each fold and compute the area under the curve. Is this result consistent with the results you obtained using the logistic regression model you implemented?

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

auc_list = []

kf = KFold(n_splits=5, shuffle=True, random_state=42)
fold_number = 1
for train_index, val_index in kf.split(X_logistic):
    X_train, X_val = X_logistic[train_index], X_logistic[val_index]
    y_train, y_val = y_logistic[train_index], y_logistic[val_index]
    
    model = LogisticRegression()
    
    model.fit(X_train, y_train)
    
    y_prob = model.predict_proba(X_val)[:, 1]
    fpr, tpr, _ = roc_curve(y_val, y_prob)
    
    roc_auc = auc(fpr, tpr)
    
    auc_list.append(roc_auc)
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - Fold {fold_number}')
    plt.legend(loc="lower right")
    plt.show()
    
    fold_number += 1

for i, auc_score in enumerate(auc_list):
    print(f"Fold {i+1} AUC:", auc_score)

avg_auc = np.mean(auc_list)
std_auc = np.std(auc_list)

print("Average AUC:", avg_auc)
print("Standard Deviation AUC:", std_auc)
```
